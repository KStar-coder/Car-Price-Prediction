```python
df = pd.read_csv('car data.csv') #converting into DataFrame
print(df)
```

```python
#non numeric
# Car_Name, Fuel_Type, Seller_Type, Transmission

print(df.Fuel_Type.value_counts())
```

```python
print(df.Transmission.value_counts())
```

```python
# encoding converting possible values to 0 and 1

df.replace({'Fuel_Type': {'Petrol': 0, 'Diesel':1, 'CNG' : 2 }}, inplace  = True)
df
```

```python
df.replace( {'Seller_Type' : { 'Dealer':0, 'Individual':1}},inplace = True)
df
```

```python
df.replace( {'Transmission': {'Manual':0 , 'Automatic': 1}},inplace = True)
df
```

```python
df.info()
```

```python
#splitting x = independent variables
x = df.drop(['Car_Name' , 'Selling_Price'], axis = 1)
```

```python
print(x)
```

```python
y = df['Selling_Price'] # Dependent value
y
```

```python
from sklearn.model_selection import train_test_split #for testing data 
```

```python
x_train , x_test, y_train , y_test = train_test_split(x,y, test_size = 0.2, random_state =2 ) #random state is used to control underfitting and overfitting
x_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression #importing the model
from sklearn import metrics # for checking accuracy and error correction of the model
```

```python
#Load the model

lr = LinearRegression()
```

```python
lr
```

```python
lr.fit(x_train, y_train) # training data is fed to the model
```

```python
trainPrediction = lr.predict(x_train)
testPrediction = lr.predict(x_test)
```

```python
trainPrediction
```

```python
testPrediction
```

```python
# R squared value
errorTrain = metrics.r2_score(y_train , trainPrediction)
errorTrain
```

```python
errorTest = metrics.r2_score(y_test , testPrediction)
errorTest
```

```python
plt.scatter(trainPrediction, y_train) #scatter plot
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
```

```python
plt.scatter(testPrediction, y_test) #scatter plot
plt.xlabel('Actual Price tested')
plt.ylabel('Predicted Price tested')
```
